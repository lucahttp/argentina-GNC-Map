```python
from pyproj import Transformer
import json

transformer = Transformer.from_crs("EPSG:3857", "EPSG:4326")


# print(bbcitabblin)
# convert str to json
# import ast
# encodedUnicode = ast.literal_eval(json.dumps(bbcitabblin, ensure_ascii=False))
# print(encodedUnicode)


def formatAndSave(inputjson,filename="test.json"):
    inputjson = inputjson.replace("{success:true,rows:", '{"success":true,"rows":')
    inputjson = inputjson.replace("'", '"')
    json_data = json.loads(inputjson)

    print(json_data)


    for i in range(len(json_data['rows'])):
        x = json_data['rows'][i]['x']
        y = json_data['rows'][i]['y']
        optimus = transformer.transform(x,y)
        json_data['rows'][i]['latitude'] = optimus[0]
        json_data['rows'][i]['longitude'] = optimus[1]
        


    print(json_data)

    with open(filename, 'w', encoding='utf-8') as f:
        # Serialize the data and write it to the file
        json.dump(json_data, f, ensure_ascii=False)
        # json.dump(json_data, f)
    return True

```
